```python



with open ('D:/lab2/深度学习课程设计/crnn-pytorch-master/train_history/mobileVit_rnn_3_valid', 'rb') as fp:
    temp1 = pickle.load(fp)
    
with open ('D:/lab2/深度学习课程设计/crnn-pytorch-master/train_history/mobileVit_rnn_valid', 'rb') as fp:
    temp2 = pickle.load(fp)
    
with open ('D:/lab2/深度学习课程设计/crnn-pytorch-master/train_history/crnn_1_valid', 'rb') as fp:
    temp3 = pickle.load(fp)
    
temp=[temp1,temp2,temp3]
```

```python
val_loss_list=[]
val_accuracy_list = []
for model in temp:
    loss =[]
    acc = []
    for i in range(10):
        loss.append(model[i]['loss'])
        acc.append(model[i]['acc'])
    val_loss_list.append(loss)
    val_accuracy_list.append(acc)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(val_accuracy_list[0], label='CRNN with MobileVit block' , color = 'gold')
plt.plot(val_accuracy_list[1], label='MobileVit-RNN' , color = 'blue')
plt.plot(val_accuracy_list[2], label='CRNN' , color = 'lavender')
plt.xlabel('Train step')
plt.ylabel('Value')
x = [0,4000, 8000, 12000,16000,20000, 24000,28000,32000, 36000,40000]
default_x_ticks = range(len(x))
plt.title('Validation accuracy comparison on Synth90k')
plt.xticks(default_x_ticks,x )
plt.legend()
plt.show()
```
